```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
Cancer = load_breast_cancer()
Cancer.keys()
```

```python
Cancer['feature_names']
```

```python
df = pd.DataFrame(Cancer['data'], columns=Cancer['feature_names'])
df.head()
```

```python
df['target'] = Cancer['target']
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

**Splitting the Data**

```python
x = df.drop('target', axis=1)
y = df['target']
x.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,
                                                    random_state=100)
print("Training Data", x_train.shape)
print("Testing Data", x_test.shape)
```

**Model Building**

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(x_train, y_train)
```

**Generate Predictions**

```python
y_predicted_train = svm.predict(x_train)
y_predicted_test = svm.predict(x_test)
```

**Evaluation**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Training Accuracy", accuracy_score(y_train, y_predicted_train))
print("Testing Accuracy", accuracy_score(y_test, y_predicted_test))
```

```python
print(classification_report(y_train, y_predicted_train))
```

```python
cm = confusion_matrix(y_test, y_predicted_test)
cm
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100,1000],
              'gamma': [1, 0.1, 0.01, 0.001,0.0001],
              'kernel' : ['rbf']}
```

```python
grid_model = GridSearchCV(SVC(), param_grid,
                          refit=True, verbose=1)
```

```python
grid_model.fit(x_train, y_train)
```

```python
grid_model.best_params_
```

```python
grid_predictions = grid_model.predict(x_test)
```

```python
print("Testing Accuracy = ",accuracy_score(y_test, grid_predictions))
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
cm = confusion_matrix(y_test, grid_predictions)
cm
```

```python
sns.heatmap(cm, annot=True, cmap = 'viridis_r')
```
